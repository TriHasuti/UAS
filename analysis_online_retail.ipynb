```python

import pandas as pd
import numpy as np

from pathlib import Path
import matplotlib.pyplot as plt

df = pd.read_excel("/content/online_retail_II.xlsx")  # path dataset
df
```

```python
df['Description'] = df['Description'].fillna('Unknown')
```

```python
df = df.dropna(subset=['Customer ID'])
```

```python
df.isnull().sum()
```

```python
# Normalisasi nama kolom
df.columns = [c.strip().replace(" ", "_").replace("/", "_").lower() for c in df.columns]

# Cast tipe
if "invoice_date" in df.columns:
    df["invoice_date"] = pd.to_datetime(df["invoice_date"], errors="coerce")

# Hilangkan transaksi cancel/refund:
# - Quantity <= 0 atau Invoice diawali 'C'
if "invoice" in df.columns:
    is_cancel = df["invoice"].astype(str).str.startswith("C", na=False)
else:
    is_cancel = False

if "quantity" in df.columns:
    invalid_qty = df["quantity"] <= 0
else:
    invalid_qty = False

df = df[~(is_cancel | invalid_qty)].copy()

# Feature: total_price
df["total_price"] = df.get("quantity", 0) * df.get("price", 0.0)

print("Ukuran data (bersih):", df.shape)
df.head()
```

```python
# KPI dasar
total_orders = df["invoice"].nunique() if "invoice" in df.columns else np.nan
total_customers = df["customer_id"].nunique() if "customer_id" in df.columns else np.nan
total_products = df["stockcode"].nunique() if "stockcode" in df.columns else np.nan
total_revenue = df["total_price"].sum()

print({
    "total_orders": total_orders,
    "total_customers": total_customers,
    "total_products": total_products,
    "total_revenue": round(float(total_revenue), 2)
})

# Top produk berdasarkan quantity dan revenue
top_qty = (df.groupby(["stockcode","description"], dropna=False)["quantity"]
             .sum().sort_values(ascending=False).head(10).reset_index())

top_rev = (df.groupby(["stockcode","description"], dropna=False)["total_price"]
             .sum().sort_values(ascending=False).head(10).reset_index())

top_countries = (df.groupby("country")["invoice"]
                   .nunique().sort_values(ascending=False).head(10).reset_index())

top_qty, top_rev, top_countries
```

```python
# Analisis waktu bulanan
dt = df.set_index("invoicedate").sort_index()
monthly_revenue = dt["total_price"].resample("MS").sum()  # Month start frequency
monthly_orders = dt["invoice"].resample("MS").nunique()

monthly_summary = pd.DataFrame({
    "revenue": monthly_revenue,
    "orders": monthly_orders
}).dropna(how="all")

monthly_summary
```

```python
# Plot: Revenue bulanan
plt.figure()
monthly_summary["revenue"].plot(title="Monthly Revenue")
plt.xlabel("Bulan")
plt.ylabel("Revenue")
plt.show()
```

```python
# Plot: Top 10 produk by revenue
plt.figure()
top_rev_plot = monthly_summary  # dummy to keep structure clear
top_rev_df = (df.groupby(["stockcode","description"])["total_price"]
                .sum().sort_values(ascending=False).head(10).reset_index())
top_rev_df.plot(x="description", y="total_price", kind="bar", title="Top 10 Produk (Revenue)")
plt.xlabel("Produk")
plt.ylabel("Revenue")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Plot: Top 10 negara by #orders
plt.figure()
top_countries_df = (df.groupby("country")["invoice"].nunique()
                      .sort_values(ascending=False).head(10).reset_index())
top_countries_df.plot(x="country", y="invoice", kind="bar", title="Top 10 Negara (Jumlah Order)")
plt.xlabel("Negara")
plt.ylabel("Jumlah Order")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# RFM
snapshot_date = df["invoicedate"].max() + pd.Timedelta(days=1)

rfm = df.groupby("customer_id").agg({
    "invoicedate": lambda x: (snapshot_date - x.max()).days,  # Recency (days)
    "invoice": "nunique",                                     # Frequency (# invoices)
    "total_price": "sum"                                      # Monetary
}).rename(columns={"invoicedate":"Recency","invoice":"Frequency","total_price":"Monetary"})

# Skoring kuantil (4 terbaik)
rfm["R_score"] = pd.qcut(rfm["Recency"], 4, labels=[4,3,2,1])  # lebih kecil recency lebih baik
rfm["F_score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 4, labels=[1,2,3,4])
rfm["M_score"] = pd.qcut(rfm["Monetary"].rank(method="first"), 4, labels=[1,2,3,4])
rfm["RFM_Segment"] = rfm["R_score"].astype(str) + rfm["F_score"].astype(str) + rfm["M_score"].astype(str)
rfm["RFM_Score"] = rfm[["R_score","F_score","M_score"]].astype(int).sum(axis=1)

rfm.head()
```

```python
# Simpan ringkasan ke file CSV (untuk folder output/)
out_dir = Path("/content/output")
out_dir.mkdir(parents=True, exist_ok=True)

# Tabel ringkasan
top_qty.to_csv(out_dir / "top_products_by_quantity.csv", index=False)
top_rev.to_csv(out_dir / "top_products_by_revenue.csv", index=False)
top_countries.to_csv(out_dir / "top_countries_by_orders.csv", index=False)
monthly_summary.to_csv(out_dir / "monthly_summary.csv")

# RFM
rfm.to_csv(out_dir / "rfm_scores.csv")

"Outputs tersimpan di folder output/."
```
